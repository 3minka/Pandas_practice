```python
# Embarked 탑승한 지역에 따른 평균 나이는 어떻게 될까?

# df.groupby('Sex')['Age'].mean()
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
# Cabin, Embarked 
df['Cabin'].str[0].value_counts()
df['Cabin'] = df['Cabin'].fillna('C00')
```

```python
df['Embarked'].value_counts()
df['Embarked'].fillna('S', inplace=True)
```

```python
df['Age'].value_counts() # 나이를 grouping해서 데이터를 들여다보면 또 보이는 것이 있겠죠.

# 10~20대, 30~40대, 고령자
df['Age'] <= 20, "20s"
df['Age'] <= 40, "40s"
df['Age'] > 40, "60s"
```

```python
def age_group(age):
    if age <= 20:
        group = 'Student'
    elif age <= 40:
        group = 'Adult'
    else:
        group = 'senior'
    return group
```

```python
df['Age_Group'] = df['Age'].apply(age_group)

# 나이의 그룹에 따른 생존 확률
df.groupby('Age_Group')['Survived'].mean()
```

```python
# 60대 이상의 고령자 분들은 생존 확률이 어떻게 되셨을까?
df_age_high = df[df['Age'] >= 60][['Survived', 'Age', 'Sex', 'Fare', 'Embarked']]
df_age_high.describe()
```
